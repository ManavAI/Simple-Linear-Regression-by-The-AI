* distance=(yi - y^i)^2 i from 1 to n 

* n = no of students

* y = actual value of the dependent variable

* y^ = predicted value of the dependent variable


now this distance is the error let it e


* e=(yi - y^i)^2 i from 1 to n


we know that,

* y^ = mx + b


put in e

* e=(yi - mxi - b)^2 i from 1 to n


for minimize the error we do differentialtion


1) differentiation with respect to b

* d(e)/d(b) = 0


* after performing differentiation you reach to,

# b = y(bar) - m * x(bar)


2) differentiation with respect to m


* after performing differentiation you reach to,

# m = (yi - y(bar)) (xi - x(bar)) / (xi - x(bar))^2

* where i form 1 to n


let's do this with the help of code,

```python
# It is my regression model named as AI_Linear_Regression

class AI_Linear_Regression:

    def __init__(self):
        self.coef_=None
        self.intercept_=None

    def fit(self,x,y):
        
        num=0
        den=0
        
        for i in range(x.shape[0]):
            num = num + ((y[i]-y.mean())*(x[i]-x.mean()))
            den = den + ((x[i]-x.mean())*(x[i]-x.mean()))
            
        self.coef_ = num / den
        self.intercept_ = y.mean() - (self.coef_ * x.mean())
        

    def predict(self,x):
        
        y = (self.coef_ * x) + self.intercept_
        return y
    
```
