Linear Regression
* Linear Regression is the Machine Learning technique in which we try predict the value of desired perameter from given independent perameters.
* let,
        y = a1x1 + a2x2 + a3x3 + c

        here, x1,x2,x3 are the independent variable 
              a1,a2,a3 are the cofficients
              y is the predicted value

        in linear regression our model try to predict a1,a2,a3 such a way that the result(y) should near or equal to actual_result(y).
        in other way, The model learns these coefficients during the training process using an optimization algorithm (such as gradient descent) to minimize the error between the predicted and actual values.

Types:-
1) Simple Linear Regression
    -one input one output

    CGPA	         LPA
    (independent)    (dependent)

    9	             15
    8	             10
    10	             9
    9.95	         10
    8.5	             13
    
2) Multiple Linear Regression
    -many input one output

    CGPA	        12th Marks(%)	Gender	        LPA
    (independent)	(independent)	(independent)	(dependent)
    
    9	            90	            Male	        15
    8	            80	            Female	        10
    10	            85	            Male	        9
    9.95	        95	            Female	        10
    8.5	            82	            Female	        13
    
3) Polynomial Linear Regression

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/kaggle/input/simple-linear-regression/placement.csv")
#df.head()
print(df)
```

in 200 rows * 2 columns  cgpa is idependent perameter and package is dependent perameter

```python
plt.scatter(df['cgpa'],df['package'],color='g')
plt.xlabel('CGPA',size=15,color='y')
plt.ylabel('LPA',size=15,color='y')
plt.title('CGPA v/s LPA',size=20,color='r')
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

DataFrame.iloc[x,y] returns xth row of yth column of our DataFame df

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

same for y_train and y_test

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression as lr
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
value1=lr.predict(x_test.iloc[0].values.reshape(1,1))
value2=lr.predict(x_test.iloc[1].values.reshape(1,1))
value3=lr.predict(x_test.iloc[2].values.reshape(1,1))
print(x_test.iloc[0]," -> ",value1)
print(x_test.iloc[1]," -> ",value2)
print(x_test.iloc[2]," -> ",value3)
```

* here, for cgpa 8.58 actual value of lpa is 4.10 and predicted is 3.89
*       for cgpa 7.15 actual value of lpa is 3.49 and predicted is 3.09
*       for cgpa 5.88 actual value of lpa is 2.08 and predicted is 2.38

here,
* it is simple linear regression so,
* equation is y = mx + c
* What is internally done?
>from training data, model predict appropriate value of m and c whare m is known as weight and c is known as bius. let m=m1 ans c=b1. Now when new data point come let x1 then our model predict LPA from this simple formula

# y = m1x1 + b1
let's see this,

```python
plt.scatter(x_train,y_train,color='green')
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('CGPA',size=15,color='y')
plt.ylabel('LPA',size=15,color='y')
plt.title('CGPA v/s LPA',size=20,color='red')
```

1. This red line shows in the graph is the best fit line for our model
> Best fit line is the line predicted by the model via prediction of slope and the intercept or bius which passes closely from our data points.

let's see what values of slope and bius predicted by our model for our DataFrame.

```python
m=lr.coef_
b=lr.intercept_
print('m= ',m,' & b=',b)
```

* here you can see for the value of m and b I use lr object.
* It means lr is the line but we also see lr is the output of out LinearRegression model.
# lr=LinearRegression()
* means LinearRegression model only predict a line to solve our problem
* means based of this line based on slope and bius it predict the value for the given x as y=mx+b
